```python
%%sql
SELECT *
  FROM facts
LIMIT 5
```

### Begin Page 4: Summary Statistics

```python
%%sql
SELECT MIN(population) AS min_population,
    MAX(population) AS max_population,
    MIN(population_growth) AS min_pop_growth,
    MAX(population_growth) AS max_pop_growth
  FROM facts;
```

### End Page 4: Summary Statistics

### Begin Page 5: Exploring Outliers

```python
%%sql
SELECT Name, population_growth
  FROM facts
 WHERE population_growth = (SELECT MIN(population_growth) AS min_pop_growth
                              FROM facts)
    OR population_growth = (SELECT MAX(population_growth)
                              FROM facts);
```

### End Page 5: Exploring Outliers

### Begin Page 6: Exploring Average Population and Area

```python
%%sql
SELECT MIN(population) AS min_population,
    MAX(population) AS max_population,
    MIN(population_growth) AS min_pop_growth,
    MAX(population_growth) AS max_pop_growth,
    AVG(population) AS avg_population,
    AVG(area) AS avg_area
  FROM facts
 WHERE Name <> 'World'
```

### End Page 6: Exploring Average Population and Area

### Begin Page 7: Finding Densely Populated Countries

```python
%%sql
SELECT Name, population
  FROM facts
 WHERE Name <> 'World'
   AND (population > (SELECT AVG(population) FROM facts WHERE Name <> 'World')
    OR population > (SELECT AVG(area) FROM facts WHERE Name <> 'World'))
ORDER BY Name
```

### End Page 7: Finding Densely Populated Countries
